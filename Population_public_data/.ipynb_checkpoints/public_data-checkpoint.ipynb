```python
#      

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('./age.csv', encoding='cp949')

indexs = df.columns[3:].tolist()

ages = ''

age = []
index = []

for row in df.values:
    if row[0] == '   (4146356000)':
        ages = row[3:]

for a in ages:
    age.append(int(a))

for b in indexs:
    index.append(b[11:])

fig,ax = plt.subplots()
x = 1920 / fig.dpi
y = 1080 / fig.dpi 
fig.set_figwidth(x)
fig.set_figheight(y)
ax.plot(index, age)
plt.rc('font', size = 12) #     
plt.xticks(np.arange(0,101,1), rotation=90) # x  , 
plt.yticks(np.arange(0,1001,50))
plt.rcParams['font.family'] = 'NanumGothic'
plt.xlabel('Ages') # x 
plt.ylabel('Population') # y 
plt.title('Avg_Temperature') #   
plt.grid(True) #     

plt.show(ax)
```

```python
# input   

import pandas as pd
import matplotlib.pyplot as plt 
import numpy as np
import csv

f = open('age.csv', encoding='cp949')
data =csv.reader(f)
next(data)

input_data = input()

populations = []

for id, va in enumerate(data):
    if input_data in va[0]:
        populations = va[3:]

x = np.arange(101)
y = [int(i.replace(',','')) for i in populations]

plt.plot(x, y)
plt.show()
```

```python
#  ,        

import pandas as pd
import matplotlib.pyplot as plt 
import csv
import numpy as np

f = open('gender.csv', encoding='cp949')
data = csv.reader(f)
next(data)

m = []
f = []

for v in data:
    if '' in v[0]:
        m = v[3:104]
        f = v[106:]

x = np.arange(0, 101)
y = [-int(i.replace(',','')) for i in m]
f1 = [int(a.replace(',','')) for a in f]

plt.barh(x, y, color='b', label ='Male')
plt.barh(x, f1, color='r', label ='Female')
plt.xlabel('')
plt.ylabel('')
plt.legend()
plt.rcParams['font.family'] = 'NanumGothic'
plt.show()
```

```python
# input     

import pandas as pd
import matplotlib.pyplot as plt 
import numpy as np
import csv

f = open('gender.csv', encoding='cp949')
data =csv.reader(f)
next(data)

input_data = input()

m = []
f = []

for id, va in enumerate(data):
    if input_data in va[0]:
        m = v[3:104]
        f = v[106:]

x = np.arange(101)
y = [-int(i.replace(',','')) for i in m]
z = [int(i.replace(',','')) for i in f]
f1 = [int(a) for a in z]

plt.barh(x, y, color='b', label ='Male')
plt.barh(x, f1, color='r', label ='Female')
plt.xlabel('')
plt.ylabel('')
plt.legend()
plt.rcParams['font.family'] = 'NanumGothic'
plt.show()
```
