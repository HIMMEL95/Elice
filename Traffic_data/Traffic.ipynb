# 유임 승차 비율이 가장 높은 역은? 

```python
import csv
import matplotlib.pyplot as plt
import numpy as np

f = open('subwayfee.csv', encoding='cp949')
data = csv.reader(f)
next(data)

max_station = ''
max_ = 0

for v in data:
    if int(v[4]) + int(v[6]) >= 100000:
        rate = int(v[4]) / (int(v[4]) + int(v[6]))
        if rate > max_:
            max_ = rate
            max_station = v[3]

print('유임 승차 비율(',round(max_,2),'%)이 가장 높은 역은', max_station,'입니다')
```

# 모든 역의 유무임 승하차 비율은 어떻게 될까

```python
import csv
import matplotlib.pyplot as plt
import numpy as np

f = open('subwayfee.csv', encoding='cp949')
data = csv.reader(f)
next(data)

label = ['유임승차', '유임하차', '무임승차', '무임하차']

population1 = []
population2 = []
population3 = []
population4 = []

for row in data:
    population1.append(int(row[4]))
    population2.append(int(row[5]))
    population3.append(int(row[6]))
    population4.append(int(row[7]))
       
population1 = sum(population1) / 598
population2 = sum(population2) / 598
population3 = sum(population3) / 598
population4 = sum(population4) / 598

population = []

population.append(population1)
population.append(population2)
population.append(population3)
population.append(population4)

plt.pie(population, labels=label, autopct='%.1f%%')
plt.rcParams['font.family'] = 'NanumGothic'
plt.show()
```

# 출근 시간대 사람들이 가장 많이 타는 역은 어디일까

```python
import csv 
import matplotlib.pyplot as plt
import numpy as np

f = open('subwaytime.csv', encoding = 'cp949')
data = csv.reader(f)
next(data)
next(data)

time_7 = []
time_8 = []
time_9 = []

label = []

for row in data:
    row[4:] = map(int, map(lambda x:x.replace(',',''), row[4:]))
    time_7.append(row[10])
    time_8.append(row[12])
    time_9.append(row[14])
    label.append(row[3])

time_7 = np.array(time_7)
time_8 = np.array(time_8)
time_9 = np.array(time_9)

result = time_7 + time_8 + time_9

mx = result.max()

x = np.arange(598)

plt.figure(figsize=(100,40))
plt.bar(x, result)
plt.rc('font', size=10)
plt.xticks(x, label, rotation=90)
plt.rc('font', family='NanumGothic')
plt.show()

mx = result.max()

mx_station = label[39]

print('최대 승차인원은', mx,'이고, 해당 역은',mx_station,'입니다.')
```
